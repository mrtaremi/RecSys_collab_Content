```python
# تو مرحله اول جمع اوردی داده ها ما اول باید فیلم هایی که کاربر ما دیده رو مبنای فیلتر قرار بدیم
# یعنی بر اساس فیلم های کاربر قراره فیلم های دیگه ای بهش پیشنهاد داده بشه
# پس از دل جدول امتیاز ها ما همه کاربرانی که حداقل یک فیلم مشترک با یوزر فرضی ما دارند رو جدا میکنیم
# و بقیه کاربر ها مورد استفاده قرار نمیگیره
us = dr[dr['movieId'].isin(dfu['movieId'].tolist())]
us
```

```python
# لیست به دست اماده در مرحله به این صورت است که از بین تمام یوزر ها
# ما یوزر هایی رو انتخاب کردیم که حداقل یک فیلم مشترک دارند
# پس باید در مرحله اول اونو بر اساس ایدی کاربر تقسیم بندی کنیم
usg = us.groupby(['userId'])
```

```python
# از جدول به دست اماده چه اطلاعاتی میفهمیم؟
# میفهمیم که مثلا کاربر ۲۳۰۹۵ از بین ۷ فیلم کاربر ورودی ما حداقل ۳ فیلم مشترک دارد
# این ععد ممکن است حداقل ۱ فیلم باشد با حداکثر هر ۷ فیلم باشد
usg.get_group(23095)
```

```python
# دقیق تر بخوام بگم بعد از تقسیم بندی ما یک ارایه داریم که اعضای این ارایه شامل یک تاپل هستند
# و هر تاپل شامل ۲ عضو میباشد
# عضو اول این تاپل ایدی کاربر هستش
# و عضو دوم این ارایه یک دیتا فریم از همان فیلم های مشترک

# حالا ما قرار این ارایه اصلی رو مرتب کنیم
# بر اسا چی؟ بر اساس همون عضو دوم تاپل که میشه دیتا فریم ما
# که برای این کارم از تابع سورت و تابع خطی لامبدا استفاده میکنیم
usg = sorted(usg,  key=lambda x: len(x[1]), reverse=True)
```

### Collaborate Based Recommender Systems

```python
# خوب نوبت اونه که ما ماتریس شباهتمون رو بسازیم
# شباهت هر کاربر که حداقل یک فیلم مشترک دارد و بر اساس نمره ای که به اون فیلم داده


dis = []
# خوب همون جور که گفته شد ارايه ما شامل تاپل های ۲ عضوی هستن 
# که عضو اول ایدی و عضو دوم فیلم های مشترک
# برای همین از حلقه ای استفاده شده که ۲ ورودی چرخشی دارد
for name, group in usg:
  # در این مرحله باید جدول گروه مرتب شود
  # دلیل این کار هم برای این هستش که قراره فاصله دو کاربر رو حساب کنیدپس باید فیچر های یکسانی
  # مورد مقایسه قرار بگیری، مثلا نمره فیلم الف در مقابل نمره فیلم الف ولی از دید کاربر دیگر
  # برای این کار هم میتونیم از کد زیر استفاده کنیم هم از سلول قبلی
  # group = group.sort_values(by='movieId')

  # همنجور که گفته شد باید جدول کاربر ما هم مرتب بشه تا فیلم به صورت برابر در مقابل هم قرار بگیریند
  # برای این منظور جدول کاربر هم باید مرتب بشه
  dfu = dfu.sort_values(by='movieId')


  # خوب همنجور که گفته شد باید فاصله دو کاربر با هم مورد مقایسه قرار بگمپس باید فیچر های برابری رو مقایسه قرار دارند
  # مثلا اگه کاربر شماره ۱۲۰ فقط ۳ فیلم مشترک با ما دارد
  # پس باید فقط همان ۳ فیلم مشترک مورد مورد مقایسه قرار گیرد
  # پس جدول کربرمون رو در هر مرحله محدود میکنیم به کاربر های مشترک
  temp_df = dfu[dfu['movieId'].isin(group['movieId'].tolist())]

  # خوب همه چی امادست
  # کاربر ما فیلم های ۱و۲و۳و۴و۵و۶و۷ رو دیده
  # ولی یوزر ما فقط در فیلم های ۳و۶و۷ مشتکر است

  # پس نمره فیلم های ۳و۶و۷ از کاربر ورودی و کاربر همکار در هر مرحله دریافت میشه
  tmpRatingInput = temp_df['score'].tolist()
  tmpGroupList = group['rating'].tolist()


  # مشکل اول اینجاست که اگر کاربری فقط ۱ فیلم مشترک داشته باشد نمیشود فاصله ماتریسی اونو بگیریم
  # پس تو این مرحله تمام کاربر هایی که فقط ۱ یک فیلم مشترک رو دارن حذف میکنیم.
  if (len(tmpGroupList) == 1):
    continue
  else:
    # اون کاربر هایی که بیشتر از ۱ یک فیلم مشترک دارم رو بر اساس متد پوآسون ازش ماتریس شباهت میگیریم
    # این کارو میتونیم با متد های زیر هم انجام بدیم
    # euclidean
    # sqeucli
    # cosine
    # correla
    # hamming
    # jensens
    PersonDistans = ss.pearsonr(tmpRatingInput,tmpGroupList)[0]
    dis.append([name , PersonDistans])
```

```python
# بعد از ساخت ماتریس شباهت باید اونو مرتب و اصلاح کنیم
# مرحله اول:
# پوآسون در بعضی حالت ها مقدار غیر عددی خروجی میدهد
# در این مرحله همه این مقادیر رو با صفر جا به جا میکنیم

# مرحله دوم؛
# ماتریس رو تبدیل به دیتافریم میکنیم و نام برای ستون هاش انتخاب میکنیم

# مرحله سوم:
# تمام کاربرانی که فاصله ماتریسی شون زیر صفر هستش رو حذف میکنیم
# چون قرار نیس کسانی که هیچ گونه شباهتی با ما ندارند رو مورد بررسی قرار دهیم
# دلیل بعدی هم اینکه در مرحله های جلو تر ما یک تقسیم انجام میدهیم
# که این مقدار در مخرج قرار میگیره
# اگه در مخرج صفر قرار بگیره مقدار ما برابر میشه با بینهایت که مورد قبول ما نیست
dis = pd.DataFrame(dis).fillna(0)
dis.columns = ['userId','similarity']
dis = dis[dis['similarity']>0]
dis
```

```python
# دیتافریم به دست امده حاصل کاربرانی هست که بیشترین شباهت رو به ما دارن
# پس کافیه اونو بر اسا میزان شباهت مرتب کنیم
topSimilar=dis.sort_values(by='similarity', ascending=False)
topSimilar
```

```python
# حالا که بعد از کلی سختی همت کردیم و ماتریس شباهتمون رو پیدا کردیم
# کافیه تمام فیلم هایی که اون کاربر ها دیدن رو به دست بیاریم
# هدف ما هم همین بود، بر اساس فیلم های مشترک و امتیازاتی که داده داده شده میزان شباهت رو به دست بیاریمبعد که میزان شباهت به دست اومد باید 
# فیلم هایی که کاربر های مشابه ما دیدن رو به ما معرفی کنه

# پس با تابع مرج لیست کاربر هامونو بهش میدهیم
# و در ورودی دوم لیست تمام فیلم های دیده شده توسط کاربران یا همون دیتا فریم امتیازات
# که برای اینکه دیتافریم منسجمی داشته باشیم کلید اصلی رو اینجا یوزر ایدی میدهیم که نمرات فیلم کاربر همراه با میزان شباهت و ایدی فیلم در یک ستون داشته باشیم
topUsersRating=topSimilar.merge(dr, left_on='userId', right_on='userId', how='inner')
topUsersRating
```

```python
# در این مرحله میزان شباهت اون کاربر با کاربر ورودی ضربدر نمره که کاربر به فیلم داده شده میکنیم و در یک ستون دیگر ذخیره میکنیم
topUsersRating['weighted'] = topUsersRating['similarity']*topUsersRating['rating']
topUsersRating
```

```python
# هدف ما معرفی فیلم جدید هست، پس دوباره با مشخصه ایدی فیلم دیتافریم رو دسته بندی میکنیم
# هر دسته به چه معنا؟
# هر دسته  به معنای یک فیلمی است که کاربر ممکنه بهش پیشنهاد داده بشه
# چون کاربر های متشابه بهش اون فیلم هارو دیدن،
# هر رکورد از این دسته بندی یک فیلد به نام شباهت داره
# یک فیلد هم به اسم وزن داره که حاصل شباهت ضربدر نمره کاربر میباشد
# پس با جمع تمام این موارد میزان شباهت و وزن تمام کاربر هارو در یک رکورد نگه میداریم
tempTopUsersRating = topUsersRating.groupby('movieId').sum()[['similarity','weighted']]

# دیتاست به دست اومده رو با تغییر نام ستون ها اصلاح میکنیم
tempTopUsersRating.columns = ['sum_similarity','sum_weight']
tempTopUsersRating
```

```python
# خب کافیه میزان وزن اون فیلم رو تقسیم بر میزان شباهت ها کنیم
# دلیل این کارم اینکه یک نرمال سازی روی فیچر هامون داشته باشیم
# و فیلمی که کاربر کمتری اونو دیده ولی نمرات بالا تری داره
# در مقابل فیلمی که کاربر بیشتری اونو دیده و نمرات کمتری داره تو با یک معیار سنجیده بشه
rd = pd.DataFrame()
rd['recommendation score'] = tempTopUsersRating['sum_weight']/tempTopUsersRating['sum_similarity']
rd['movieId'] = tempTopUsersRating.index
rd
```

```python
# جدول نهایی میزان پیشنهاد هر فیلم به کاربر ورودی ما هستش
# که باید مرتب سازی بر اساس همین نمره میشه با تاپ این لیست دسترسی داشت
rd = rd.sort_values(by='recommendation score', ascending=False)
rd
```

```python
# در اخر هم کافیه بر روی دیتا فریم فیلم های اصلی
# مرتب کنیم بر اساس فیلم های پیشنهاد شده
# مانند سیستم پیشنهادی محتوا
RecSysCollaborative = df.reindex(rd['movieId'].tolist()).dropna()
```
